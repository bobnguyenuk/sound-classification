```python
features=np.load("20features.npy")
features.shape
```

```python
labels=np.hstack((np.zeros((100,),dtype=int),np.ones((100,),dtype=int)))
```

```python
print('Class labels:', np.unique(labels))
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.svm import SVC
from sklearn import metrics
for i in range(1, 16):
    X_train, X_test, y_train, y_test = train_test_split(features, labels , test_size=.2, random_state=i)
    #Create the SVM model
    clf = SVC(C=100, kernel = 'rbf')

    #Fit the model for the data
    clf.fit(X_train, y_train)

    #Make the prediction
    y_pred = clf.predict(X_test)
    
    print ('--- Case', i, '---')
#     # Model Accuracy: 
#     print("Accuracy: {:.2f}".format(metrics.accuracy_score(y_test, y_pred)))
#     # Model Precision: 
#     print("Precision: {:.2f}".format(metrics.precision_score(y_test, y_pred)))
#     # Model Recall: 
#     print("Sensitivity (Recall): {:.2f}".format(metrics.recall_score(y_test, y_pred)))
    
#     # Confusion matrix
#     print(metrics.confusion_matrix(y_test,y_pred))
#     print(metrics.classification_report(y_test,y_pred))
    
    # Cross-validation
    accuracies = cross_val_score(estimator = clf, X = X_train, y = y_train, cv = 5)
    print("5 fold cross-validation -> Accuracy: {:.2f}".format(accuracies.mean()))
```
